```python
sns.scatterplot(data=data,x='sepal_length',y='sepal_width',hue='species')
```

```python
#elbow technique
d_rnge = range(1,10)
s=[]
for i in d_rnge:
  a=KMeans(n_clusters=i)
  a.fit(data[['petal_length','petal_width']])
  s.append(a.inertia_)
```

```python
plt.xlabel('d_rnge')
plt.ylabel('squared errors')
plt.plot(d_rnge,s,color='r')
```

```python
s
```

```python
#kmean algo
k=KMeans(n_clusters=3,random_state=0)
y=k.fit_predict(data[['petal_length','petal_width']])
y
```

```python
data['cluster']=y
data.head(150)
```

```python
#accuracy
from sklearn.metrics import confusion_matrix
a=confusion_matrix(data.species,data.cluster)
a
```

```python
a=data.species
pred=data.cluster
c=confusion_matrix(a,pred)
cls=['setosa','versicolor','virginica']
plt.imshow(c,interpolation='nearest',cmap=plt.cm.Blues)
plt.title('confusion matrix',color='g',fontdict={'fontsize': 18, 'fontweight': 'bold', 'color': 'green','fontname':'sans-serif'})
plt.colorbar()
tm=np.arange(len(cls))
plt.xticks(tm,cls)
plt.yticks(tm,cls)
for i in range(len(cls)):
  for j in range(len(cls)):
    plt.text(j,i,str(c[i][j]), ha='center',va='center',color='white')

plt.xlabel('Predicted table')
plt.ylabel('true tabel')
plt.show()
```
